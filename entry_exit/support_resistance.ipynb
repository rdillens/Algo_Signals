```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
ticker = yf.Ticker('MSFT')

hist = ticker.history(period='30d', interval='1d')

df_original = pd.DataFrame(hist)
df_original.head()
```

```python
df_traditional = df_original.copy()

# Traditional S/R
df_traditional['PP'] = (df_traditional['High'].shift(1) + df_traditional['Low'].shift(1) + df_traditional['Close'].shift(1)) / 3
df_traditional['R1'] = df_traditional['PP'] * 2 - df_traditional['Low'].shift(1)
df_traditional['R2'] = df_traditional['PP'] + (df_traditional['High'].shift(1) - df_traditional['Low'].shift(1))
df_traditional['R3'] = df_traditional['PP'] * 2 + (df_traditional['High'].shift(1) - 2 * df_traditional['Low'].shift(1))
df_traditional['R4'] = df_traditional['PP'] * 3 + (df_traditional['High'].shift(1) - 3 * df_traditional['Low'].shift(1))
df_traditional['R5'] = df_traditional['PP'] * 4 + (df_traditional['High'].shift(1) - 4 * df_traditional['Low'].shift(1))
df_traditional['S1'] = df_traditional['PP'] * 2 - df_traditional['High'].shift(1)
df_traditional['S2'] = df_traditional['PP'] - (df_traditional['High'].shift(1) - df_traditional['Low'].shift(1))
df_traditional['S3'] = df_traditional['PP'] * 2 - (2 * df_traditional['High'].shift(1) - df_traditional['Low'].shift(1))
df_traditional['S4'] = df_traditional['PP'] * 3 - (3 * df_traditional['High'].shift(1) - df_traditional['Low'].shift(1))
df_traditional['S5'] = df_traditional['PP'] * 4 - (4 * df_traditional['High'].shift(1) - df_traditional['Low'].shift(1))

df_traditional.tail()
```

```python
df_fibonacci = df_original.copy()


# Fibonacci S/R
df_fibonacci['PP'] = (df_fibonacci['High'].shift(1) + df_fibonacci['Low'].shift(1) + df_fibonacci['Close'].shift(1)) / 3
df_fibonacci['R1'] = df_fibonacci['PP'] + 0.382 * (df_fibonacci['High'].shift(1) - df_fibonacci['Low'].shift(1))
df_fibonacci['S1'] = df_fibonacci['PP'] - 0.382 * (df_fibonacci['High'].shift(1) - df_fibonacci['Low'].shift(1))
df_fibonacci['R2'] = df_fibonacci['PP'] + 0.618 * (df_fibonacci['High'].shift(1) - df_fibonacci['Low'].shift(1))
df_fibonacci['S2'] = df_fibonacci['PP'] - 0.618 * (df_fibonacci['High'].shift(1) - df_fibonacci['Low'].shift(1))
df_fibonacci['R3'] = df_fibonacci['PP'] + (df_fibonacci['High'].shift(1) - df_fibonacci['Low'].shift(1))
df_fibonacci['S3'] = df_fibonacci['PP'] - (df_fibonacci['High'].shift(1) - df_fibonacci['Low'].shift(1))
```

```python
df_fibonacci.tail()
```

```python
df_woodie = df_original.copy()

# Woodie S/R
df_woodie['PP'] = (df_woodie['High'].shift(1) + df_woodie['Low'].shift(1) + 2 * df_woodie['Open']) / 4
df_woodie['R1'] = 2 * df_woodie['PP'] - df_woodie['Low'].shift(1)
df_woodie['S1'] = 2 * df_woodie['PP'] - df_woodie['High'].shift(1)
df_woodie['R2'] = df_woodie['PP'] + (df_woodie['High'].shift(1) - df_woodie['Low'].shift(1))
df_woodie['S2'] = df_woodie['PP'] - (df_woodie['High'].shift(1) - df_woodie['Low'].shift(1))
df_woodie['R3'] =  df_woodie['High'].shift(1) + 2 * (df_woodie['PP'] -  df_woodie['Low'].shift(1))
df_woodie['S3'] =  df_woodie['Low'].shift(1) - 2 * (df_woodie['High'].shift(1) - df_woodie['PP'])
df_woodie['R4'] = df_woodie['R3'] + (df_woodie['High'].shift(1) - df_woodie['Low'].shift(1))
df_woodie['S4'] = df_woodie['S3'] - (df_woodie['High'].shift(1) - df_woodie['Low'].shift(1))
```

```python
df_woodie.tail()
```

```python
df_classic = df_original.copy()

# Classic S/R
df_classic['PP'] = (df_classic['High'].shift(1) + df_classic['Low'].shift(1) + df_classic['Close'].shift(1)) / 3
df_classic['R1'] = 2 * df_classic['PP'] - df_classic['Low'].shift(1)
df_classic['S1'] = 2 * df_classic['PP'] - df_classic['High'].shift(1)
df_classic['R2'] = df_classic['PP'] + (df_classic['High'].shift(1) - df_classic['Low'].shift(1))
df_classic['S2'] = df_classic['PP'] - (df_classic['High'].shift(1) - df_classic['Low'].shift(1))
df_classic['R3'] = df_classic['PP'] + 2 * (df_classic['High'].shift(1) - df_classic['Low'].shift(1))
df_classic['S3'] = df_classic['PP'] - 2 * (df_classic['High'].shift(1) - df_classic['Low'].shift(1))
df_classic['R4'] = df_classic['PP'] + 3 * (df_classic['High'].shift(1) - df_classic['Low'].shift(1))
df_classic['S4'] = df_classic['PP'] - 3 * (df_classic['High'].shift(1) - df_classic['Low'].shift(1))
```
